```python
# словари для категориальных признаков
d2n = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'June':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12}
n2d = dict((v,k) for k,v in d2n.items())

vt2n = {'Returning_Visitor':1, 'New_Visitor':2, 'Other':3}
n2vt = dict((v,k) for k,v in vt2n.items())
```

```python
df.Month = df.Month.map(d2n)
df.VisitorType = df.VisitorType.map(vt2n)
df.Weekend = df.Weekend * 1
df.Revenue = df.Revenue * 1
```

```python
# колонка общего времени на сайте
df['Duration'] = df['Administrative_Duration'] + df['Informational_Duration'] + df['ProductRelated_Duration']
```

```python
# сплит данных
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df.drop(columns=['Revenue']), df['Revenue'], test_size=0.2)
```

```python
# оверсемплинг для баланса
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=0)
x_ros,y_ros = ros.fit_resample(df.drop(columns=['Revenue']), df['Revenue'])
```

```python
# сплит новых данных
from sklearn.model_selection import train_test_split

ros_X_train, ros_X_test, ros_y_train, ros_y_test = train_test_split(
    x_ros, y_ros, test_size=0.2)
```

```python
# классификатор
from sklearn.ensemble import AdaBoostClassifier

clf = AdaBoostClassifier(n_estimators=400, random_state=0)
clf.fit(ros_X_train, ros_y_train)

# проверка на оригинальном наборе
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
print(accuracy_score(y_test,clf.predict(X_test)))
print(recall_score(y_test,clf.predict(X_test)))
print(precision_score(y_test,clf.predict(X_test)))
print(f1_score(y_test,clf.predict(X_test)))
```

```python
# результаты без оверсемплинга
clf.fit(X_train, y_train)
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
print(accuracy_score(y_test,clf.predict(X_test)))
print(recall_score(y_test,clf.predict(X_test)))
print(precision_score(y_test,clf.predict(X_test)))
print(f1_score(y_test,clf.predict(X_test)))
```

```python
# пайплайн

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer

duration_sum = lambda x: x['Administrative Duration'] + x['Informational Duration'] + x['Product Related Duration']

transformer = ColumnTransformer(
    transformers=[
        ('month', FunctionTransformer(lambda x: x.map(d2n)), ['Month']),
        ('visitor_type', FunctionTransformer(lambda x: x.map(vt2n)), ['VisitorType']),
        ('weekend', FunctionTransformer(lambda x: x * 1), ['Weekend']),
        ('revenue', FunctionTransformer(lambda x: x * 1), ['Revenue']),
        ('total_duration', FunctionTransformer(duration_sum), ['Administrative Duration', 'Informational Duration', 'Product Related Duration'])
    ],
    remainder='passthrough'
)

pipeline = Pipeline(steps=[
    ('transformer', transformer),
    ('classifier', clf)
])
```
